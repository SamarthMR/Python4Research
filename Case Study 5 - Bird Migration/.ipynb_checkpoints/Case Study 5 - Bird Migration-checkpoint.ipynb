**load data**

```python
import pandas as pd

birddata = pd.read_csv("bird_tracking.csv")
birddata.info()
```

```python
birddata.head()
```

**Simple Data Visualizations**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

ix = birddata.bird_name == "Eric"

x, y = birddata.longitude[ix], birddata.latitude[ix]

plt.figure(figsize = (7, 7))
plt.plot(x, y, '.')
```

```python
bird_names = pd.unique(birddata.bird_name)
bird_names
```

```python
plt.figure(figsize = (7, 7))

for bird_name in bird_names:
    ix = birddata.bird_name == bird_name
    x, y = birddata.longitude[ix], birddata.latitude[ix]
    plt.plot(x, y, '.', label=bird_name)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.legend(loc=4)
plt.savefig("3traj.pdf")
```

**Examining Flight Speed**

```python
ix = birddata.bird_name == "Eric"
speed = birddata.speed_2d[ix]

# number of nan
np.sum(np.isnan(speed))
```

```python
ind = np.isnan(speed)

plt.figure(figsize=(8,4))
plt.hist(speed[~ind], bins=np.linspace(0, 30, 20), normed=True)
plt.xlabel("2D speed $(m/s)$")
plt.ylabel("Frequency")
plt.savefig("hist.pdf");
```

```python
# make histogram using pandas - we don't have to deal with nan explicitly

birddata.speed_2d.plot(kind='hist', range=[0, 30])
plt.xlabel("2D speed")
plt.savefig("pd_hist.pdf")
```

**Using Datetime**

```python
birddata.date_time.head()
```

```python
import datetime

datetime.datetime.today()
```

```python
time_1 = datetime.datetime.today()
```

```python
time_2 = datetime.datetime.today()
time_2 - time_1
```

```python
date_str = birddata.date_time[0]
date_str[:-3]
```

```python
datetime.datetime.strptime(date_str[:-3], "%Y-%m-%d %H:%M:%S")
```

```python
timestamps = []
for k in range(len(birddata)):
    timestamps.append(datetime.datetime.strptime(birddata.date_time.iloc[k][:-3], "%Y-%m-%d %H:%M:%S"))

timestamps[:3]
```

```python
birddata["timestamp"] = pd.Series(timestamps, index = birddata.index)
birddata.head()
```

```python
birddata.timestamp[4] - birddata.timestamp[3] 
```

```python
# time elapsed from the beginning
times = birddata.timestamp[birddata.bird_name == "Eric"]
elapsed_time = [time - times[0] for time in times]

elapsed_time[:5]
```

```python
# calculate how many days have passed
elapsed_time[1000] / datetime.timedelta(days=1)
```

```python
# calculate how many hours have passed
elapsed_time[1000] / datetime.timedelta(hours=1)
```

```python
plt.plot(np.array(elapsed_time) / datetime.timedelta(days=1))
plt.xlabel("Observation")
plt.ylabel("Elapsed time (days)")
plt.savefig("timeplot.pdf")
```

**Calculating Daily Mean Speed**

```python
times = birddata.timestamp[birddata.bird_name == "Eric"]
elapsed_time = [time - times[0] for time in times]
days = np.array(elapsed_time) // datetime.timedelta(days=1)

birddata["day"] = pd.Series(days)
birddata.head()
```

```python
eric_data = birddata[birddata.bird_name == 'Eric']
daily_mean_speed = eric_data.groupby('day').speed_2d.mean()

plt.figure(figsize=(8, 6))
plt.plot(daily_mean_speed)
plt.xlabel("Day")
plt.ylabel("Mean speed (m/s)")
```

**Using the Cartopy Library**

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
proj = ccrs.Mercator()

plt.figure(figsize=(10,10))
ax = plt.axes(projection=proj)
ax.set_extent((-25.0, 20.0, 52.0, 10.0))
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')

for name in bird_names:
    ix = birddata.bird_name == name
    x, y = birddata.longitude[ix], birddata.latitude[ix]
    ax.plot(x, y, '.', transform=ccrs.Geodetic(), label=name)
    
plt.legend(loc=2)
plt.savefig("map.pdf")
```
