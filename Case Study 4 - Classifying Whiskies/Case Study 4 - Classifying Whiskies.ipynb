## Loading and Inspecting Data

```python
import numpy as np
```

```python
whisky = pd.read_csv("whiskies.txt")
whisky["Region"] = pd.read_csv("regions.txt")
```

```python
whisky.head()
```

```python
whisky.iloc[0:10] # index data frame by location
```

```python
whisky.iloc[5:10, 0:5]
```

```python
whisky.columns
```

```python
flavors = whisky.iloc[:, 2:14]
flavors.head()
```

## Exploring Correlations

```python
corr_flavors = pd.DataFrame.corr(flavors)
corr_flavors
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.pcolor(corr_flavors)
plt.colorbar()
plt.savefig("corr_flavors.pdf")
```

```python
corr_whisky = pd.DataFrame.corr(flavors.transpose())
plt.figure(figsize=(10, 10))
plt.pcolor(corr_whisky)
plt.colorbar()
plt.axis("tight")
plt.savefig("corr_whisky.pdf")
```

## Clustering Whiskies By Flavor Profile

```python
from sklearn.cluster.bicluster import SpectralCoclustering
```

```python
model = SpectralCoclustering(n_clusters=6, random_state=0)
model.fit(corr_whisky)
model.rows_
```

```python
np.sum(model.rows_, axis=1)
```

```python
np.sum(model.rows_, axis=0)
```

```python
model.row_labels_
```

```python
whisky.index
```

## Comparing Correlation Matrices

```python
whisky['Group'] = pd.Series(model.row_labels_, index=whisky.index) # extract group labels from model and append to whisky table
whisky = whisky.ix[np.argsort(model.row_labels_)] # reorder rows by group labels
whisky = whisky.reset_index(drop=True) # reset index of the DataFrame
whisky
```

```python
correlations = pd.DataFrame.corr(whisky.iloc[:, 2:14].transpose())
correlations = np.array(correlations)

plt.figure(figsize = (14, 7))
plt.subplot(121)
plt.pcolor(corr_whisky)
plt.title("Originial")
plt.axis("tight")

plt.subplot(122)
plt.pcolor(correlations)
plt.title("Rearranged")
plt.axis("tight")
plt.savefig("correlations.pdf")
```
